```python
# Install & Import Libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Create a simulated dataset similar to UCI's student performance
data = {
    'math_score': [50, 67, 90, 30, 75, 88, 95, 45, 60, 80],
    'reading_score': [52, 70, 85, 40, 78, 90, 94, 50, 65, 82],
    'writing_score': [48, 65, 88, 35, 72, 87, 92, 47, 60, 79],
    'gender': ['F', 'M', 'F', 'M', 'F', 'F', 'M', 'F', 'M', 'M']
}
df = pd.DataFrame(data)
```

```python
# Check for nulls and basic info
print(" Data Info:")
print(df.info())
print("\n Missing values:\n", df.isnull().sum())

# Add total and average scores
df['total_score'] = df[['math_score', 'reading_score', 'writing_score']].sum(axis=1)
df['average_score'] = df['total_score'] / 3
```

```python
# Gender distribution
sns.countplot(x='gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

```python
# Score comparison by gender
sns.boxplot(x='gender', y='average_score', data=df)
plt.title("Average Score by Gender")
plt.show()
```

```python
# Histogram of total scores
plt.hist(df['total_score'], bins=5, color='lightblue', edgecolor='black')
plt.title("Histogram of Total Scores")
plt.xlabel("Score")
plt.ylabel("Frequency")
plt.show()
```

```python
# Normalize scores to 01 scale
df['normalized_score'] = (df['total_score'] - df['total_score'].min()) / (df['total_score'].max() - df['total_score'].min())

print("\n Normalized Scores:\n", df[['total_score', 'normalized_score']])
```

```python
# final output
print("\n Summary Stats:")
print(df.describe())

print("\n Top Performer:")
print(df[df['total_score'] == df['total_score'].max()])
```
